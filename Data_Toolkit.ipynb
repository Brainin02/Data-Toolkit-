1. What is NumPy, and why is it widely used in Python?
ans. NumPy (Numerical Python) is a powerful open-source library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these data structures efficiently.

NumPy is widely used for :

Performance: NumPy is significantly faster than Python lists because it uses optimized C and Fortran code.

Memory Efficiency: It consumes less memory by storing data in contiguous blocks.

Multi-dimensional Arrays: The ndarray object allows for easy manipulation of multi-dimensional data.

Broadcasting: Enables element-wise operations without writing loops, making code more concise and efficient.

Vectorized Operations: Eliminates the need for explicit loops, improving execution speed.

Interoperability: Works well with other libraries like Pandas, SciPy, Matplotlib, and TensorFlow.

Scientific and Machine Learning Applications: It is a fundamental building block for data science and AI applications.

2. How does broadcasting work in NumPy?
ans. Broadcasting is a powerful feature in NumPy that allows arrays of different shapes to be combined in arithmetic operations without making explicit copies or resizing. Instead of performing element-wise operations only on arrays of the same shape, NumPy automatically expands smaller arrays to match the shape of larger ones.

6. What are the differences between NumPy arrays and Python lists?

ans. oth NumPy arrays and Python lists can store collections of elements, but they have significant differences in terms of performance, memory usage, and functionality.

Speed: NumPy is Faster

NumPy arrays are significantly faster than Python lists because they use C-based implementations and vectorized operations.

Python lists store elements as separate objects in memory, making them slower in computations.

Memory Efficiency: NumPy Uses Less Memory

NumPy arrays use contiguous memory storage, leading to lower memory consumption.

Python lists store references to objects, increasing memory usage.

Functionality: NumPy Has More Mathematical Operations

NumPy provides built-in mathematical operations like mean, sum, sqrt, and matrix operations.

Python lists require manual loops or using external functions like sum().

Data Type Consistency: NumPy Uses Homogeneous Data Types

NumPy arrays store elements of the same data type (homogeneous).

Python lists can hold multiple data types (heterogeneous) but at the cost of performance.

Broadcasting: NumPy Supports Vectorized Operations

NumPy supports broadcasting, meaning element-wise operations can be performed without loops.

 Python lists require explicit looping for such operations.

```python
#A Use Pandas to load a CSV file and display its first 5 rows
# Step 1: Import Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Create Sample Data
data = {
    "Name": ["brainin", "Bobo", "luffy ", "tanu", "zoro", "Frank", "naaruto", "ivan", "messi", "ney"],
    "Age": [25, 30, 27, 35, 29, np.nan, 40, 23, 31, 28],  # One missing value
    "Salary": [50000, 60000, 55000, 75000, 72000, 48000, 80000, np.nan, 69000, 58000],  # One missing value
    "Department": ["HR", "IT", "IT", "Finance", "HR", "IT", "Finance", "IT", "HR", "Finance"],
    "JobRole": ["Manager", "Developer", "Developer", "Analyst", "Manager", "Analyst", "Manager", "Developer", "Analyst", "Analyst"],
    "Date": ["2023-01-15", "2023-02-10", "2023-03-05", "2023-01-25", "2023-02-20", "2023-04-10", "2023-05-15", "2023-03-30", "2023-06-01", "2023-07-20"]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].median(), inplace=True)
df.drop_duplicates(inplace=True)
df["Salary"] = pd.to_numeric(df["Salary"])
print("Cleaned DataFrame:\n", df)
plt.figure(figsize=(8, 5))
sns.histplot(df["Salary"], bins=10, kde=True, color="blue")
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"], hue=df["Department"], palette="viridis", s=100)
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
dept_salary = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary by Department:\n", dept_salary)
pivot_table = df.pivot_table(values="Salary", index="Department", columns="JobRole", aggfunc="mean")
print("\nPivot Table (Salary by Department & Job Role):\n", pivot_table)

```

```python
#Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    "X": np.random.randint(1, 100, 50),
    "Y": np.random.randint(1, 100, 50),
    "Z": np.random.randint(1, 100, 50),
    "Category": np.random.choice(["A", "B", "C"], 50)
})

fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Category",
                     title="3D Scatter Plot Example",
                     labels={"X": "X Axis", "Y": "Y Axis", "Z": "Z Axis"},
                     size_max=10, opacity=0.8)

fig.show()
```
